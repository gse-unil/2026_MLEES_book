You are given station records as a list of tuples `(code, name, elevation_m)`:

## Exercise 2 — classify and count

Given `readings_celsius = [-2.3, 6.1, 9.4, 15.0, 22.0, -0.5]`, classify each value as "freezing" (< 0), "cold" (< 10), "mild" (< 20), or "warm", and count how many fall in each category. Return the counts as a dict.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
readings_celsius = [-2.3, 6.1, 9.4, 15.0, 22.0, -0.5]
counts = {"freezing": 0, "cold": 0, "mild": 0, "warm": 0}
for t in readings_celsius:
    if t < 0.0:
        counts["freezing"] += 1
    elif t < 10.0:
        counts["cold"] += 1
    elif t < 20.0:
        counts["mild"] += 1
    else:
        counts["warm"] += 1
print(counts)
```

Given `dates = ["Mon", "Tue", "Wed", "Thu", "Fri"]` and `readings_celsius = [6.1, 7.4, 5.9, 8.2, 6.8]`, use `enumerate` to find the index of the warmest reading, then print the matching date and temperature.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
dates = ["Mon", "Tue", "Wed", "Thu", "Fri"]
readings_celsius = [6.1, 7.4, 5.9, 8.2, 6.8]
warmest_i = 0
for i, temp in enumerate(readings_celsius):
    if temp > readings_celsius[warmest_i]:
        warmest_i = i
print(dates[warmest_i], readings_celsius[warmest_i])
```

## Exercise 4 — means and a filter, with comprehensions

Recreate a `stations` dict where each code maps to a dict containing `readings_celsius`:

```python
{"JFJ": {"readings_celsius": [-2.3, -1.1, 0.4, 1.2, -0.8]},
 "BAS": {"readings_celsius": [6.1, 7.4, 5.9, 8.2, 6.8]},
 "LUG": {"readings_celsius": [9.4, 10.2, 8.8, 11.1, 9.9]}}
```

With a dict comprehension, build `means = {code: mean_temp}`. Then, with a list comprehension, build the list of codes whose mean is below 0 °C.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
stations = {
    "JFJ": {"readings_celsius": [-2.3, -1.1, 0.4, 1.2, -0.8]},
    "BAS": {"readings_celsius": [6.1, 7.4, 5.9, 8.2, 6.8]},
    "LUG": {"readings_celsius": [9.4, 10.2, 8.8, 11.1, 9.9]},
}
means = {code: sum(v["readings_celsius"]) / len(v["readings_celsius"])
         for code, v in stations.items()}
cold_codes = [code for code, m in means.items() if m < 0.0]
print({code: round(m, 2) for code, m in means.items()})
print(cold_codes)
```

## Exercise 5 — a summary function with validation

Write a pure, type-hinted function `summary(readings)` that returns a tuple `(minimum, mean, maximum)`. Raise `ValueError` if `readings` is empty. Call it inside a `try/except` on a valid list and on an empty list.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
def summary(readings: list[float]) -> tuple[float, float, float]:
    if not readings:
        raise ValueError("readings is empty")
    return (min(readings), sum(readings) / len(readings), max(readings))

for data in ([6.1, 7.4, 5.9], []):
    try:
        print(summary(data))
    except ValueError as err:
        print("error:", err)
```

## Exercise 6 — fix the shared-default bug

An assistant wrote this function; it is correct the first time but wrong on later calls:

```python
def collect_above(values, threshold, out=[]):
    for v in values:
        if v > threshold:
            out.append(v)
    return out
```

In one comment, explain the bug, then rewrite it so each call starts fresh. Demonstrate with two separate calls.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
# out=[] is created once at definition and shared across calls, so results accumulate
def collect_above(values, threshold, out=None):
    if out is None:
        out = []
    for v in values:
        if v > threshold:
            out.append(v)
    return out

print(collect_above([1.0, 5.0, 9.0], 4.0))   # [5.0, 9.0]
print(collect_above([2.0, 8.0], 4.0))         # [8.0], not contaminated
```

Two stations report daily quality flags:

```python
jfj_flags = ["ok", "ok", "suspect", "ok", "missing"]
```

Using sets, print the flags that appear at both stations, and the flags that appear at JFJ but not at BAS.

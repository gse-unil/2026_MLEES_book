## Exercise 3 — types and division

Given `total_mm = 17` and `n_days = 5`, compute the mean daily precipitation two
ways: with `/` and with `//`. Print both results and their types, and explain in a
comment why they differ.


````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
total_mm = 17
n_days = 5
mean_true = total_mm / n_days       # float: 3.4
mean_floor = total_mm // n_days     # int: 3 (fractional part discarded)
print(mean_true, type(mean_true).__name__)
print(mean_floor, type(mean_floor).__name__)
# '/' always yields a float; '//' floors to a whole number, losing the remainder.
```

## Exercise 4 — filter and average

The raw array `[12.1, 13.4, -9999.0, 14.2, -9999.0, 11.8]` contains the sensor
error code `-9999.0`. Build a clean list with a loop and an `if`, then compute its
mean with a custom function.


````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
raw = [12.1, 13.4, -9999.0, 14.2, -9999.0, 11.8]
clean = []
for value in raw:
    if value != -9999.0:
        clean.append(value)

def mean(values):
    return sum(values) / len(values)

print(clean, round(mean(clean), 2))
```

## Exercise 6 — threshold count

Given the daily mean temperatures `[1.5, -2.0, 4.0, -0.5, 8.0, -3.1]`, count how
many days were below freezing (strictly less than 0 °C) and print the count.


````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python

    if t < 0:
```

## Exercise 7 — code review

The script below was generated to remove the `-9999.0` error codes. Decide whether
it is correct. Test it on `[12.1, -9999.0, -9999.0, 14.2]`, explain what happens,
and rewrite it so it is always correct.

```python
raw = [12.1, -9999.0, -9999.0, 14.2]
for value in raw:
    if value == -9999.0:
        raw.remove(value)
print(raw)
```


````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
The script mutates the list while iterating it. Removing an element shifts later
elements down while the loop index advances, so an element can be skipped — with
two adjacent error codes, one survives. The output is `[12.1, -9999.0, 14.2]`,
silently wrong. Build a new list instead:

```python
raw = [12.1, -9999.0, -9999.0, 14.2]
clean = [v for v in raw if v != -9999.0]
print(clean)        # [12.1, 14.2]
```

## Exercise 2 — unit conversions

A temperature is recorded as 18.5 °C. Compute it in kelvin and in degrees Fahrenheit. Separately, a discharge of 4.7 m³ s⁻¹ drains a catchment of area 250 km². Express the specific discharge in mm per day (use: 1 m³ s⁻¹ spread over 1 km² for one day equals 86.4 mm).

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
temp_celsius = 18.5
temp_kelvin = temp_celsius + 273.15
temp_fahrenheit = temp_celsius * 9 / 5 + 32

discharge_m3s = 4.7
area_km2 = 250.0
specific_discharge_mm_day = discharge_m3s / area_km2 * 86.4
print(f"{specific_discharge_mm_day:.3f} mm/day")
```

## Exercise 5 — the lexicographic trap

You are given temperature readings as strings: `["8.4", "9.5", "10.2", "7.1", "9.9"]`. First print the result of `max()` applied directly to the list, then compute and print the true numerical maximum. Add a comment explaining why they differ.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
readings_str = ["8.4", "9.5", "10.2", "7.1", "9.9"]
print(max(readings_str))            # '9.9' — lexicographic maximum

hottest_celsius = float(readings_str[0])
for r in readings_str:
    if float(r) > hottest_celsius:
        hottest_celsius = float(r)
print(hottest_celsius)              # 10.2

# they differ because string comparison is character-by-character:
# '9.9' > '10.2' is True since '9' comes after '1'
```

## Exercise 6 — round-trip through a file

Write the readings `[-2.3, -1.1, 0.4, 1.2, -0.8, -3.1, -1.9]` (°C) to `my_week.txt`, one per line, using `pathlib`. Read them back, cast to float, and print the mean and the range (max − min) to two decimals. Use `math.isclose` to confirm that the write-then-read round trip leaves the mean unchanged.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
from pathlib import Path
import math

readings_celsius = [-2.3, -1.1, 0.4, 1.2, -0.8, -3.1, -1.9]
path = Path("my_week.txt")
path.write_text("\n".join(str(v) for v in readings_celsius) + "\n", encoding="utf-8")

recovered = []
for line in path.read_text(encoding="utf-8").splitlines():
    recovered.append(float(line))

mean_celsius = sum(recovered) / len(recovered)
range_celsius = max(recovered) - min(recovered)
print(f"mean = {mean_celsius:.2f} °C, range = {range_celsius:.2f} °C")
print(math.isclose(mean_celsius, sum(readings_celsius) / len(readings_celsius)))
```

## Exercise 7 — never trust `==` on floats

Show that `0.1 + 0.2 == 0.3` is `False`, print the actual value of `0.1 + 0.2`, and then write a correct equality check using `math.isclose`.

````python
# Your solution here
````

:::{admonition} Solution
:class: note dropdown
```python
import math
print(0.1 + 0.2 == 0.3)              # False
print(0.1 + 0.2)                     # 0.30000000000000004
print(math.isclose(0.1 + 0.2, 0.3))  # True
```
